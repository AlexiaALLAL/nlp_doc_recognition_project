```python
def get_labels_from_annotation(html_file):
    labels = []
    words = []
    inside = False
    for i, word in enumerate(html_file.split()):
        if "<a>" in word:
            inside = True
        elif "</a>" in word:
            inside = False
        else:
            if inside:
                labels.append(1)
                # if labels[-1] == 0:
                #     labels.append(1)
                # else:
                #     labels.append(2)
            else:
                labels.append(0)
            words.append(word)
    return labels, words
```

```python
def read_annoted_data(file_path):
    # the target has been annotated by hand
    # the words to be extracted are between <a> and </a> tags
    list_labels = []
    list_words = []
    for folder in os.listdir(file_path):
        for file in os.listdir(file_path + folder):
            with open(file_path + folder + "/" + file, 'r') as f:
                html = f.read()
                # keep only sections where data-title="ABROGATION DE DISPOSITIONS ANTERIEURES"
                # html = re.findall(r'data-title="ABROGATION DE DISPOSITIONS ANTERIEURES".*?</section>', html, re.DOTALL)
                html = re.findall(r'<header class="dsr-header">.*?</header>', html, re.DOTALL)
                if html:
                    # print(html[0])
                    labels, words = get_labels_from_annotation(html[0])
                    # divide the text into chunks of 100 words (for the tokenizer to work)
                    for i in range(0, len(labels), 250):
                        list_labels.append(labels[i:i+100])
                        list_words.append(words[i:i+100])
    return list_labels, list_words
```

```python
def build_data(file_path):
    list_labels, list_words = read_annoted_data(file_path)
    # data has to be in the form of a list of dictionaries
    # {'id': 'i',
    # 'tokens': ['word1', word2', ...],
    # 'ner_tags': [0, 0, 1, 2, 0, ...]}
    data = []
    for i in range(len(list_labels)):
        # print(i, list_labels[i], list_words[i])
        data.append({'id': str(i), 'tokens': list_words[i], 'ner_tags': list_labels[i]})

    # separate the data into train, validation and test sets
    random.seed(42)
    random.shuffle(data)

    train_size = int(0.8 * len(data))
    val_size = int(0.1 * len(data))

    train_data = data[:train_size]
    val_data = data[train_size:train_size+val_size]
    test_data = data[train_size+val_size:]

    train_dataset = Dataset.from_dict({
        'id': [d['id'] for d in train_data],
        'tokens': [d['tokens'] for d in train_data],
        'ner_tags': [d['ner_tags'] for d in train_data]
    })
    val_dataset = Dataset.from_dict({
        'id': [d['id'] for d in val_data],
        'tokens': [d['tokens'] for d in val_data],
        'ner_tags': [d['ner_tags'] for d in val_data]
    })
    test_dataset = Dataset.from_dict({
        'id': [d['id'] for d in test_data],
        'tokens': [d['tokens'] for d in test_data],
        'ner_tags': [d['ner_tags'] for d in test_data]
    })

    data = DatasetDict({
        'train': train_dataset,
        'validation': val_dataset,
        'test': test_dataset
    })

    return data

file_path = "./data_4/data_annoted/"
data = build_data(file_path)
data
```

```python
example = data["train"][0]
tokenized_input = tokenizer(example["tokens"], is_split_into_words=True)
tokens = tokenizer.convert_ids_to_tokens(tokenized_input["input_ids"])
print(tokens)
```

```python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)

    labels = []
    for i, label in enumerate(examples[f"ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)  # Map tokens to their respective word.
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:  # Set the special tokens to -100.
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:  # Only label the first token of a given word.
                label_ids.append(label[word_idx])
            else:
                label_ids.append(-100)
            previous_word_idx = word_idx
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```

```python
tokenized_data = data.map(tokenize_and_align_labels, batched=True)
print(tokenized_data["train"][0])
```

```python
from transformers import DataCollatorForTokenClassification

data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)
```

```python
import evaluate

seqeval = evaluate.load("seqeval")
```

```python
import numpy as np

labels = [label_list[i] for i in example[f"ner_tags"]]


def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    true_predictions = [
        [label_list[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [label_list[l] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    results = seqeval.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
```

```python
training_args = TrainingArguments(
    output_dir="ner-model",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=1,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_data["train"],
    eval_dataset=tokenized_data["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
def get_labels_from_annotation(html_file, nb_labels=2):
    labels = []
    words = []
    inside = False
    for i, word in enumerate(html_file.split()):
        if "<a>" in word:
            inside = True
        elif "</a>" in word:
            inside = False
        else:
            if inside:
                if nb_labels == 2:
                    labels.append(1)
                else:
                  if labels[-1] == 0:
                      labels.append(1)
                  else:
                      labels.append(2)
            else:
                labels.append(0)
            words.append(word)
    return labels, words
```

```python
def get_labels_from_annotation(html_file, nb_labels=2):
    labels = []
    words = []
    inside = False
    for i, word in enumerate(html_file.split()):
        if "<a>" in word:
            inside = True
        elif "</a>" in word:
            inside = False
        else:
            if inside:
                if nb_labels == 2:
                    labels.append(1)
                else:
                    if labels[-1] == 0:
                        labels.append(1)
                    else:
                        labels.append(2)
            else:
                labels.append(0)
            words.append(word)
    return labels, words
```

```python
def read_annoted_data(file_path, nb_labels=2):
    # the target has been annotated by hand
    # the words to be extracted are between <a> and </a> tags
    list_labels = []
    list_words = []
    for folder in os.listdir(file_path):
        for file in os.listdir(file_path + folder):
            with open(file_path + folder + "/" + file, 'r') as f:
                # print(file_path + folder + "/" + file)
                html = f.read()
                # keep only sections where data-title="ABROGATION DE DISPOSITIONS ANTERIEURES"
                # html = re.findall(r'data-title="ABROGATION DE DISPOSITIONS ANTERIEURES".*?</section>', html, re.DOTALL)
                html = re.findall(r'<header class="dsr-header">.*?</header>', html, re.DOTALL)
                if html:
                    # print(html[0])
                    labels, words = get_labels_from_annotation(html[0], nb_labels)
                    # divide the text into chunks for the tokenizer to work (max 512 for roberta)
                    for i in range(0, len(labels), 256):
                        list_labels.append(labels[i:i+100])
                        list_words.append(words[i:i+100])
    return list_labels, list_words
```

```python
def build_data(file_path):
    list_labels, list_words = read_annoted_data(file_path)
    # data has to be in the form of a list of dictionaries
    # {'id': 'i',
    # 'tokens': ['word1', word2', ...],
    # 'ner_tags': [0, 0, 1, 2, 0, ...]}
    data = []
    for i in range(len(list_labels)):
        # print(i, list_labels[i], list_words[i])
        data.append({'id': str(i), 'tokens': list_words[i], 'ner_tags': list_labels[i]})

    # separate the data into train, validation and test sets
    random.seed(42)
    random.shuffle(data)

    train_size = int(0.8 * len(data))
    val_size = int(0.1 * len(data))

    train_data = data[:train_size]
    val_data = data[train_size:train_size+val_size]
    test_data = data[train_size+val_size:]

    train_dataset = Dataset.from_dict({
        'id': [d['id'] for d in train_data],
        'tokens': [d['tokens'] for d in train_data],
        'ner_tags': [d['ner_tags'] for d in train_data]
    })
    val_dataset = Dataset.from_dict({
        'id': [d['id'] for d in val_data],
        'tokens': [d['tokens'] for d in val_data],
        'ner_tags': [d['ner_tags'] for d in val_data]
    })
    test_dataset = Dataset.from_dict({
        'id': [d['id'] for d in test_data],
        'tokens': [d['tokens'] for d in test_data],
        'ner_tags': [d['ner_tags'] for d in test_data]
    })

    data = DatasetDict({
        'train': train_dataset,
        'validation': val_dataset,
        'test': test_dataset
    })

    return data

file_path = "./data_4/annoted_regex/"
data = build_data(file_path)
data
```

```python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)

    labels = []
    for i, label in enumerate(examples[f"ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)  # Map tokens to their respective word.
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:  # Set the special tokens to -100.
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:  # Only label the first token of a given word.
                label_ids.append(label[word_idx])
            else:
                label_ids.append(-100)
            previous_word_idx = word_idx
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```

```python
def add_tags_identification(div_identification, verbose=False):
    # add <a> </a> tags around the identification
    # print(div_identification)
    # the identification starts with "Arr√™t√©" or "ARR√äT√â"
    
    normalized_text = unidecode(div_identification)

    pattern_date = r'ARRETE.*\d{2}.*\d{4}'
    found_pattern_date = re.findall(pattern_date, normalized_text, re.IGNORECASE | re.UNICODE)

    pattern_number_list = [
        r'ARRETE.*\b\d{2}/[A-Z]+/\d{3}\b',
        r'ARRETE.*\b\d{2}/[A-Z0-9]+/\d{3}\b',
        r'ARRETE.*\b\d{4} - \d{2}-\d{2}\b',
        r'ARRETE.*\b\d{4}-\d{4}-\d{2}\b',
        r'ARRETE.*\b\d{4}-\d{2}-\d{2}\b',
        r'ARRETE.*\b\d{4}-\d{2}-\d{2}-\d{3}\b',
        r'ARRETE.*\b\d{4}-\d{2}-\d{5}\b',
        r'ARRETE.*\b[A-Z]\d{2}-\d{2}-\d{2}-\d{3}\b',
        r'ARRETE.*\b\d{3}-\d{2}-\d{2}-\d{3}\b',
        r'ARRETE.*\d{4}-\d{2}-\d{2}-\d{3}\b',
        r'ARRETE.*\b\d{4}-\d{2}-\d{3}\b',
        r'ARRETE.*\b[A-Z]\d{1}-[A-Z]\d{1}-\d{2}-\d{3}\b',
        r'ARRETE.*\b\d{7}[A-Z]{2}\(\d{3}.\d{5}\)\b',
        r'ARRETE.*\b\d{4}-\d{2}\b',
        r'ARRETE.*\b\d{4}-\d{3}-[A-Z]{2}\b',
        r'ARRETE.*\b\d{4}-\d{2}-[A-Z]+\b',
        r'ARRETE.*\d{4}-\d{3}-[A-Z]+\b',
        r'ARRETE.*\b\d{7}-\d{4}\b',
        r'ARRETE.*\b\d{4}/\d{6}\b',
        r'ARRETE.*\d{4}/\d{2}/\d{2}',
        r'ARRETE.*\b\d{8}-\d{1}\b'
    ]
    found_pattern_number = []
    for i, pattern_number in enumerate(pattern_number_list):
        found_pattern_number = re.findall(pattern_number, normalized_text, re.IGNORECASE | re.UNICODE)
        if found_pattern_number:
            good_pattern_number = pattern_number
            break
    
    pattern_else_list = [
        r'ARRETE.*MISES\s+EN\s+DEMEURE\b',
        r"ARRETE.*AUTORISATION",
        r"ARRETE.*PRESCRIPTIONS\s+COMPLEMENTAIRES",
        r"ARRETE.*PREFECTORAL\s+COMPLEMENTAIRE",
        r"ARRETE.*PREFECTORAL",
        r"ARRETE"
    ]
    found_pattern_else = []
    for i, pattern_else in enumerate(pattern_else_list):
        found_pattern_else = re.findall(pattern_else, normalized_text, re.IGNORECASE | re.UNICODE)
        if found_pattern_else:
            good_pattern_else = pattern_else
            break
    
    if found_pattern_date:
        if verbose:
            print("Found date pattern", found_pattern_date[0])
        return re.sub(pattern_date, r"<a>\g<0></a>", normalized_text, flags=re.IGNORECASE | re.UNICODE)

    elif found_pattern_number:
        if verbose:
            print("Found number pattern", found_pattern_number[0])
        return re.sub(good_pattern_number, r"<a>\g<0></a>", normalized_text, flags=re.IGNORECASE | re.UNICODE)
    
    elif found_pattern_else:
        if verbose:
            print("Found else pattern", found_pattern_else[0])
        return re.sub(good_pattern_else, r"<a>\g<0></a>", normalized_text, flags=re.IGNORECASE | re.UNICODE)
    
    else:
        if verbose:
            print("No pattern found")

    
```

```python
def test_all_identification(file_path):
    """
    Test all identifications
    """

    for folder in os.listdir(file_path):
        for file in os.listdir(file_path + folder):
            with open(file_path + folder + "/" + file, 'r') as f:
                html = f.read()
                html = re.findall(r'<header class="dsr-header">.*?</header>', html, re.DOTALL)
                div_identification = extract_identification(html[0])
                # print(div_identification)
                div_identification_with_tags = add_tags_identification(div_identification, verbose=False)
                if not div_identification_with_tags:
                    print(div_identification)
                    add_tags_identification(div_identification, verbose=True)
                if div_identification_with_tags == div_identification:
                    print(div_identification)
                    add_tags_identification(div_identification, verbose=True)
                    print("No change")

file_path = "./data_4/data/"
test_all_identification(file_path)
```

```python
def add_tags_visa(div_visa, verbose=False):
    
    normalized_text = unidecode(div_visa)
    # print(normalized_text)

    pattern_name_doc_list = [
        # r'CODE DE L\'ENVIRONNEMENT',
        # r'CODE DE L ENVIRONNEMENT',
        r'code de l\'environnement',
        r'code de l environnement',
        r'code des relations entre le public et l\'administration',
        r'code des relations entre le public et l administration',
        r'code de la sant√© publique'
    ]
    pattern_relative = [
        r'(directive.*)relative',
        r'(l\'arrete.*)relatif',
        r'(arrete.*)relatif',
        r'(decret.*)relatif',
        r'(circulaire.*)relative',
        r'(rapport.*)relatif',
        r'(loi.*)relative',
        r'(l\'ordonnance.*)relative',
        r'(ordonnance.*)relative',
    ]
    pattern_date = [        
        r'l\'arrete.*\d{1,2}.*\d{4}', # l'arr√™t√© ... du ...
        r'arrete.*\d{1,2}.*\d{4}', # arr√™t√© ... du 28 f√©vrier 2017
        r'directive.*\d{1,2}.*\d{4}', # directive ... du ...
        r'rapport.*\d{1,2}.*\d{4}', # rapport du 4 ao√ªt 2021
        r'decret.*\d{1,2}.*\d{4}', # d√©cret ... du ...
        r'circulaire.*\d{1,2}.*\d{4}', # circulaire ... du ...
        r'loi.*\d{1,2}.*\d{4}', # loi ... du ...
        r'l\'ordonnance.*\d{1,2}.*\d{4}', # l'ordonnance ... du ...
        r'ordonnance.*\d{1,2}.*\d{4}', # ordonnance ... du ...
    ]
    else_pattern = [
        r'arrete.*[0-9A-Z]{2}/[0-9A-Z]+/[0-9A-Z]{3}',
        r'directive.*[0-9A-Z]{4}/[0-9A-Z]{2}/[0-9A-Z]{2}',
    ]


    found_pattern_name_doc = []
    for i, pattern_name_doc in enumerate(pattern_name_doc_list):
        found_pattern_name_doc = re.findall(pattern_name_doc, normalized_text, re.IGNORECASE | re.UNICODE)
        if found_pattern_name_doc:
            good_pattern_name_doc = pattern_name_doc
            break
    
    found_pattern_relative = []
    for i, pattern_relative in enumerate(pattern_relative):
        found_pattern_relative = re.findall(pattern_relative, normalized_text, re.IGNORECASE | re.UNICODE)
        if found_pattern_relative:
            good_pattern_relative = pattern_relative
            break
    
    found_pattern_date = []
    for i, pattern_date in enumerate(pattern_date):
        found_pattern_date = re.findall(pattern_date, normalized_text, re.IGNORECASE | re.UNICODE)
        if found_pattern_date:
            good_pattern_date = pattern_date
            break
    
    found_pattern_else = []
    for i, pattern_else in enumerate(else_pattern):
        found_pattern_else = re.findall(pattern_else, normalized_text, re.IGNORECASE | re.UNICODE)
        if found_pattern_else:
            good_pattern_else = pattern_else
            break
    
    if found_pattern_name_doc:
        if verbose:
            print("Found name_doc pattern", found_pattern_name_doc[0])
        return re.sub(good_pattern_name_doc, r"<a>\g<0></a>", normalized_text, flags= re.IGNORECASE | re.UNICODE)
    
    elif found_pattern_relative:
        if verbose:
            print("Found relative pattern", found_pattern_relative[0])
        return re.sub(good_pattern_relative, r"<a>\g<0></a>", normalized_text, flags= re.IGNORECASE | re.UNICODE)

    elif found_pattern_date:
        if verbose:
            print("Found date pattern", found_pattern_date[0])
        return re.sub(good_pattern_date, r"<a>\g<0></a>", normalized_text, flags= re.IGNORECASE | re.UNICODE)

    elif found_pattern_else:
        if verbose:
            print("Found else pattern", found_pattern_else[0])
        return re.sub(good_pattern_else, r"<a>\g<0></a>", normalized_text, flags= re.IGNORECASE | re.UNICODE)
    
    else:
        if verbose:
            print("No pattern found")
```

```python
def create_annoted_files(file_path, annoted_path):
    """
    Create new_file in the folder annoted_path with the identification and visa sections tagged
    """
    for folder in os.listdir(file_path):
        for file in os.listdir(file_path + folder):
            file_name = file_path + folder + "/" + file
            # print(file_name)
            with open(file_path + folder + "/" + file, 'r') as f:
                html_full = f.read()
                html = re.findall(r'<header class="dsr-header">.*?</header>', html_full, re.DOTALL)
                # print(html[0])
                # create new file with identification and visa sections tagged
                new_file = annoted_path + folder + "/" + file
                # create the folder if it doesn't exist
                if not os.path.exists(annoted_path + folder):
                    os.makedirs(annoted_path + folder)
                with open(new_file, 'w') as f:
                    # identification section
                    div_identification = extract_identification(html[0])
                    div_identification_with_tags = add_tags_identification(div_identification, verbose=False)
                    html_full = html_full.replace(div_identification, div_identification_with_tags)

                    # visa sections
                    div_visa_list = extract_visa(html[0])
                    if div_visa_list:
                        for div_visa in div_visa_list:
                            div_visa_with_tags = add_tags_visa(div_visa, verbose=False)
                            if div_visa_with_tags:
                                html_full = html_full.replace(div_visa, div_visa_with_tags)
                    f.write(html_full)
                    
```
